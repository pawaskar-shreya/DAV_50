AIM:Set Up a D3.js Environment, Select Elements in D3, Modify Elements in D3,Data Loading in D3

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Basic D3 Scatter Plot</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
    <style>
        .dot {
            fill: steelblue;
        }
    </style>
</head>
<body>

<h2>Basic D3 Scatter Plot</h2>
<div id="chart-container"></div>

<script>

    const data = [
        { x: 10, y: 20 },
        { x: 40, y: 90 },
        { x: 80, y: 50 },
        { x: 60, y: 70 },
        { x: 30, y: 40 }
    ];

    const width = 400;
    const height = 300;
    const margin = { top: 20, right: 20, bottom: 30, left: 40 };

    const svg = d3.select("#chart-container")
        .append("svg")
        .attr("width", width)
        .attr("height", height);

    const xScale = d3.scaleLinear()
        .domain([0, d3.max(data, d => d.x)])
        .range([margin.left, width - margin.right]);

    const yScale = d3.scaleLinear()
        .domain([0, d3.max(data, d => d.y)])
        .range([height - margin.bottom, margin.top]);

    svg.selectAll("circle")
        .data(data)
        .enter().append("circle")
        .attr("class", "dot")
        .attr("cx", d => xScale(d.x))
        .attr("cy", d => yScale(d.y))
        .attr("r", 5);

    svg.append("g")
        .attr("transform", `translate(0,${height - margin.bottom})`)
        .call(d3.axisBottom(xScale));

    svg.append("g")
        .attr("transform", `translate(${margin.left},0)`)
        .call(d3.axisLeft(yScale));
</script>

</body>
</html>
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Changing SVG Element Color with D3.js</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>

<h2>Changing SVG Element Color with D3.js</h2>
<button id="color-button">Change Rectangle Color</button>
<svg width="400" height="200">
  <rect id="myRect" x="50" y="50" width="100" height="50" fill="steelblue"></rect>
</svg>

<script>
    function changeRectColor() {
        const randomColor = randomColors();
        d3.select("#myRect")
            .transition()
            .duration(500)
            .attr("fill", randomColor);
    }

    d3.select("#color-button")
        .on("click", changeRectColor);

    function randomColors() {
        const letters = "0123456789ABCDEF";
        let color = "#";
        for (let i = 0; i < 6; i++) {
            color += letters[Math.floor(Math.random() * 16)];
        }
        return color;
    }
</script>

</body>
</html>
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Interactive Bar Chart with D3.js</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
    <style>
        .bar {
            fill: steelblue;
        }

        .bar:hover {
            fill: orange;
        }
    </style>
</head>
<body>

<h2>Interactive Bar Chart with D3.js</h2>
<div id="chart-container"></div>

<script>

    const data = [30, 86, 168, 201, 203, 125];


    const width = 420;
    const height = 300;

    const svg = d3.select("#chart-container")
        .append("svg")
        .attr("width", width)
        .attr("height", height);

    svg.selectAll("rect")
        .data(data)
        .enter().append("rect")
        .attr("class", "bar")
        .attr("x", (d, i) => i * 70)
        .attr("y", d => height - d)
        .attr("width", 65)
        .attr("height", d => d)
        .on("click", function() {
            d3.select(this).attr("fill", randomColors());
        });

    svg.selectAll("text")
        .data(data)
        .enter().append("text")
        .text(d => d)
        .attr("x", (d, i) => i * 70 + 30)
        .attr("y", d => height - d - 5)
        .style("font-size", "14px")
        .style("text-anchor", "middle");

    function randomColors() {
        const letters = "0123456789ABCDEF";
        let color = "#";
        for (let i = 0; i < 6; i++) {
            color += letters[Math.floor(Math.random() * 16)];
        }
        return color;
    }
</script>

</body>
</html>
```
